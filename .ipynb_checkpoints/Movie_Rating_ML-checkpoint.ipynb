# Movie Rating By Machine Learning

*In machine learning, a movie rating is a numerical or categorical value predicted by a model that represents the likelihood of a user enjoying a particular movie. These predictions are typically generated by analyzing patterns in user preferences, historical ratings, and various movie attributes, such as genre, cast, and director. The goal is to provide personalized recommendations to users by accurately forecasting their potential ratings for unseen movies.*



# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,MultiLabelBinarizer
from sklearn import linear_model
import plotly.express as px

import warnings
warnings.filterwarnings(category=UserWarning,action='ignore')
```

# Data Collection

```python
data=pd.read_excel('Movie_Rating.xlsx','movie_rating')

df=pd.DataFrame(data)
df.head()
```

# Data Pre-Processing

```python
df=df.drop(['Genre','Poster_Link','Series_Title','Released_Year','Overview','Director','Star1','Star2','Star3','Star4'],axis='columns')
df.head()
```

```python
df.Gross=df.Gross.fillna(np.mean(df.Gross))
df.head()
```

```python
le=LabelEncoder()
mlb = MultiLabelBinarizer()

df['Runtime'] = df['Runtime'].astype(str).str.replace(' min', '').astype(int)
df['Certificate'] = le.fit_transform(df['Certificate'].astype(str))

df
```

```python
x=df.drop(['IMDB_Rating'],axis='columns')
x=x.fillna(0)
x.head()
```

```python
y=df.IMDB_Rating
y.head()
```

# Model Training and Testing

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.9)

le=linear_model.LinearRegression()
model=le.fit(xtrain,ytrain)
model
```

```python
model.predict([[1,175,100.0,1620367	,134966411.0]])
```

```python
model.score(xtest,ytest)
```
